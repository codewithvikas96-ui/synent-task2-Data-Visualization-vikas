### Phase 1 — Data Cleaning

```python
# null values
print("Checking Null values in dataset")
chk = df.isnull().sum()
print(chk)
print(f"There are {sum(chk)} null values in the dataset")

null_pct = (sum(chk) / len(df)) * 100 
print(f"Null values pct: {null_pct}%")
```

```python
# Finding Duplicate Rows and droping them
duplicate = df.duplicated()
print(f"Duplicate rows in dataset: {sum(duplicate)}")
print(df[duplicate])
print()

df = df.drop_duplicates()
print(f"Shape of dataset after removing duplicate rows")
print(f"Rows: {df.shape[0]}")
print(f"Columns: {df.shape[1]}")
```

```python
# Detecting Outliers using IQR
numeric_cols = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

def detect_outliers_iqr(numeric_cols,df):
    outliers = {}
    for col in numeric_cols: 
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1 
        lower_fence = Q1 - 1.5 * IQR
        upper_fence = Q3 + 1.5 * IQR
        outliers[col] = df[(df[col] < lower_fence) | (df[col] > upper_fence)][col]

    return outliers

outliers = detect_outliers_iqr(numeric_cols,df)
print("Outliers per columns")
outliers
```

### Phase 2 — Preprocessing

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
```

```python
# Encode the species column into numeric labels (0, 1, 2) using LabelEncoder
encoder = LabelEncoder()
df['species_encoded'] = encoder.fit_transform(df['species'])
df.sample(10)
```

```python
# StandardScaler on the 4 numeric features 
scaler = StandardScaler()
scaled = scaler.fit_transform(df[numeric_cols])
df_scaled = pd.DataFrame(scaled, columns=numeric_cols)
df_scaled.head()
```

```python
# difference between original and scaled values
comparison = pd.concat([df[numeric_cols], df_scaled.add_suffix('_scaled')], axis=1)
comparison.head(10)
```

### Phase 3 — EDA

```python
# Mean of each feature based on species 
grp_mean = df.groupby('species').mean()
print("Mean of each feature based on species")
grp_mean
```

```python
max_plen = grp_mean[grp_mean['petal_length'] == grp_mean['petal_length'].max()]
print(f"{max_plen.index[0]} specie has max petal length {max_plen['petal_length'].round(2).iloc[0]} on average")
```

```python
# Correlation between numeric features
corr_matrix = df[numeric_cols].corr()
sns.heatmap(corr_matrix, annot=True)

plt.xticks(rotation=0)
plt.yticks(rotation=0)
plt.show()


corr_pairs = corr_matrix.unstack()
sorted_pairs = corr_pairs.sort_values(kind="quicksort", ascending=False)

strongest_pair = sorted_pairs[sorted_pairs < 1].idxmax()
max_corr_value = sorted_pairs[sorted_pairs < 1].max()

print(f"The most correlated features are {strongest_pair[0]} and {strongest_pair[1]} with correlation {max_corr_value:.2f}")
```

```python
# variance 
col_var = df[numeric_cols].var()
print("Variance of numeric feature")
print(col_var)
print()
print(f"{col_var.idxmax()} has the highest variance of {col_var.max():.2f} across the entire dataset")
```

```python
# For each species, print the min and max of petal_width using .groupby().agg().
df.groupby('species')['petal_width'].agg(['min','max'])
```

### Phase 4 — Visualizations

```python
#  Bar chart showing average petal_length for each species  
avg_pl = df.groupby('species')['petal_length'].mean()

plt.figure(figsize=(8,6))
ax = sns.barplot(x=avg_pl.index, y=avg_pl.values, palette='hot', edgecolor='black')

for container in ax.containers: 
    ax.bar_label(container, fmt='%.2f', padding=4)

plt.title("Average petal_length for each species", fontweight='bold')
plt.xlabel("Species")
plt.ylabel("Avg Petal Length (cm)")
plt.tight_layout()
plt.show()
```

```python
# Histogram for all 4 features in a single figure (2×2 grid) 
fig, axes = plt.subplots(2,2, figsize=(14,12))

axes = axes.flatten()

colors = ["firebrick", "salmon", "gold", "purple"]

for i,(ax, col, cr) in enumerate(zip(axes,numeric_cols, colors)): 
    sns.histplot(x=df[col],kde=True, alpha=0.9, ax=ax, color=cr)
    ax.set_title(f"Histogram of {col}", fontweight='bold')
    ax.legend([col])  

fig.suptitle("Histogram of all 4 numeric features", fontweight='bold', fontsize=16)

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

```python
# Scatter plot of sepal_length vs petal_length, colored by species 
sns.scatterplot(
    data=df,
    x="sepal_length",
    y="petal_length",
    hue="species",  
    palette="hot",  
    s=80,            
    edgecolor="black"
)

plt.title("Scatter plot of Sepal Length vs Petal Length", fontweight='bold')
plt.show()
```

```python
# Boxplot for all 4 features grouped by species.
fig, axes = plt.subplots(2, 2, figsize=(14, 12))
axes = axes.flatten()

colors = ["firebrick", "salmon", "gold", "purple"]

for ax, col, cr in zip(axes, numeric_cols, colors): 
    sns.boxplot(data=df, x="species", y=col, ax=ax, palette=[cr])
    ax.set_title(f"Boxplot of {col} by species", fontweight='bold')

fig.suptitle("Boxplot of all 4 numeric features grouped by species", fontweight='bold', fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

```python
# Heatmap of the correlation matrix with annotation values  
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')

plt.xticks(rotation=0)
plt.yticks(rotation=0)
plt.show()
```

```python
# CPairplot showing all feature relationships colored by species.
sns.pairplot(df, vars=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'], hue='species')
plt.show()
```
